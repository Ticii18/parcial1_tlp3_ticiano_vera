# 1. IMPORTACION DE DATASET

 1. en el primer bloque de codigo lo que hacemos es importar pandas y a la vez leemos el csv

```python
import pandas as pd

df= pd.read_csv("ventas.csv")

df.head(5)
```

# 2. Exploracion de datos
1. En este cuadro de codigo, lo que hacemos es mostrar tanto las primeras 5 filas, como las 5 ultimas.
2. Tambien mostramos los "Dtype" de cada columna y los valores nulos que hay en el csv

```python
print(f"Las primeras 5 filas {df.head(5)}")
print(f"\n Las ultimas 5 filas {df.tail(5)}")

print(f"\n {df.info()} ")

print(f"Los valores nulos son {df.isnull().sum()}")
```

# 3. Limpieza y Normalización de los Datos
1. Para el primer bloque de codigo lo que hacemos es convertir la fecha a datetime y los valores nulos, es decir los que no tienen fecha le ponemos en "0"
2. En el segundo bloque lo que hacemos es convertir todos los valores nulos a un respectivo valor, ya sea "No metodo", en metodos de pago y asi con las demas columnas.
3. Luego al "Precio_Unitario" lo valores faltantes lo que hacemos es rellenarlos con la media
4. Luego convertimos los nombres y apellidos, para que tengan la primera letra en mayusculas.

```python
fecha1=df['Fecha']=pd.to_datetime(df['Fecha'], format="%Y-%m-%d", errors="coerce")
fecha2=df['Fecha']=pd.to_datetime(df['Fecha'], format="%d-%m-%y", errors="coerce")

df['Fecha']=df['Fecha']= fecha1.combine_first(fecha2) 


df['Metodo_Pago'] = df['Metodo_Pago'].fillna(0)  
df['Metodo_Pago'] = df['Metodo_Pago'].replace(0.0, "No metodo")

df['Region'] = df['Region'].fillna(0)
df['Region'] = df['Region'].replace(0.0, "No region")

df['Producto'] = df['Producto'].fillna(0)
df['Producto'] = df['Producto'].replace(0.0, "No hay productos")

df['Total_Venta'] = df['Total_Venta'].fillna(0)

df['Precio_Unitario'].fillna(df['Precio_Unitario'].mean(), inplace=True)


df['Nombre_Cliente'] = df['Nombre_Cliente'].str.title()
df['Apellido_Cliente'] = df['Apellido_Cliente'].str.title()

print(f"{df.head(5)}")


print(f"\n Mostrar los valores nulos actualizados: \n {df.isnull().sum()} ")

```

# 4. Obtención de Estadísticas Descriptivas
1. Lo que hacemos en el bloque de codigo completo es describir todo el df, es decir el csv y tambien agrupamos por regio y metodo de pago para saber cual es total de ventas en dichas regiones.

```python

print(f"{df.describe()}")



df.groupby(['Region', 'Metodo_Pago'])['Total_Venta'].sum()
```

# 5. Visualización de los Datos con Matplotlib
1. En este caso creamos un grafico de linea, donde se puede ver el total de ventas por region.

```python
import matplotlib.pyplot as plt

ventas_region = df.groupby('Region')['Total_Venta'].sum()

ventas_region.plot(kind='line')
plt.title("Total de ventas por region")
plt.xlabel("Region")
plt.ylabel("Total de ventas")
plt.show()
```

# 6. Exportación a Archivo SQLite
1. Para terminar el parcial, lo que hacemos es exportar el csv todo modificado a una base de datos sqlite
2. Luego hacemos una consulta de la base de datos para corroborar si se exporto bien.

```python
import sqlite3 

conex= sqlite3.connect("ventas.db")
df.to_sql("VentasDB", conex, if_exists="replace",index=False )

pd.read_sql("Select * from VentasDB LIMIT 5", conex)

conex.close()
```
